# EXPLORATORY DATA ANÁLISIS (EDA)

cargamos las libreria con que vamos hacer el EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Vamos a realizar el eda del archivo de la tabla de hechos que se encuentra en el archivo de Excel de nombre homicidios, con el fin de poder darle un vistazo a nuestros datos y conocer sus características básicas. 

```python
df=pd.read_excel('homicidios.xlsx',sheet_name='HECHOS') # cargamos los datos de la table hechos
```

Lo primero que vamos a conocer es la dimensión de nuestra table de hechos la cual es de (696, 21), esto quiere decir que esta tabla tiene 696 filas y 21 columnas. 

```python
df.shape
```

En la próxima tabla vamos a observar una breve descripción de los datos que se encuentra en cada una de las columnas de la table de hechos, así como las características de los datos que están en ella es decir el tipo de dato (si son datos enteros, decimal, datatime). De esta  tabla tambien podemos deducir que las columnas Calle, Alturas, Cruce tienen valores faltantes, para saber con mayor precisión la cantidad de valores faltantes en estas columnas tenemos la siguiente tabla. La cual nos indica el numero de fila vacías por columnas.

```python
df.info()
```

En la proxima tabla podemos deducir que las columnas Alturas y Cruces tienen el mayor numero de datos faltantes por lo cual debemos considerar la posibilidad de eliminarla debido a la gran ausencia de información en ellas.

```python
df.isna().sum()
```

En la siguiente grafica podemos observar un histograma de la columna N_Victimas, en donde podemos identificar la cantidad de la cantidad de víctimas por siniestros viales. De este histograma se puede concluir que da la totalidad de siniestros o accidentes viales por lo regular tiene como resultado una persona afectada. Entiéndase como afectada que puede resultar afectación leve, afectación grave, muerta.

```python
#plt.figure(figsize=(12, 6))
plt.hist(df['N_VICTIMAS'], bins=2)  # 'bins' es el número de divisiones del histograma

plt.xticks(range(min(df['N_VICTIMAS']), max(df['N_VICTIMAS'])))
# Agregar etiquetas y título

plt.xlabel('Numero de Victimas')
plt.ylabel('Frecuencia')
plt.title('Histograma de Numero de Victimas')

# Mostrar el histograma
plt.show()
```

En el proximo histograma corresponde a la columna AAAA (años), este grafico ilustra la cantidad de accidentes vihales entre los aos de 2016 a 2021. De este grafico se puede inferir que hay una dismiucion del numero de accidende entre 2016 a 2020 con excepcion del año 2018 el numero de accidende disminuyo de un año con respecto al otro. en el año 2021 hubo nuevamente un repunte en la accidentalidad, lo cual se explica en gran medida a que en el años 2020 año hubo la pandemia del covid-19 en donde se presento el aislamento para la poblacion lo cual ocaciono que hubieran una menor cantidad de vehiculos trasitando. por lo cual no seria ideal compara el año 2021 con el año 2020.

```python
# plt.figure(figsize=(12, 6))
plt.hist(df['AAAA'], bins=6)  # 'bins' es el número de divisiones del histograma

#plt.xticks(range(min(df['AAAA']), max(df['AAAA'])+1))
# Agregar etiquetas y título

plt.xlabel('Año')
plt.ylabel('Numero De Siniestros')
plt.title('Histograma De Siniestros Vailes Por Año')

# Mostrar el histograma
plt.show()
```

En el próximo histograma el cual corresponde al número de siniestros por mes. Podemos ver de forma más clara como el numero de siniestros aumenta de forma progresiva los últimos cuatros meses del año alcanzando su pico máximo en el último mes del año.

```python
#plt.figure(figsize=(12, 6))
plt.hist(df['MM'], bins=12)  # 'bins' es el número de divisiones del histograma

plt.xticks(range(min(df['MM']), max(df['MM'])+1))
# Agregar etiquetas y título

plt.xlabel('Numero de Victimas')
plt.ylabel('Frecuencia')
plt.title('Histograma de Siniestros por Mes')

# Mostrar el histograma
plt.show()
```

En el siguiente histograma se observa el numero de siniestros viales por hora, dado este grafico podemos concluir que entre la seis y la siete de la mañana es la hora del día donde ocurre el mayor numero de siniestros

```python
 plt.figure(figsize=(12, 6))
plt.hist(df[df['HH']!="SD"]['HH'], bins=24)  # 'bins' es el número de divisiones del histograma

plt.xticks(range(min(df[df['HH']!="SD"]['HH']), max(df[df['HH']!="SD"]['HH'])+1))
# Agregar etiquetas y título

plt.xlabel('Hora Del Dia')
plt.ylabel('Numero De Siniestros')
plt.title('Histograma De Siniestros Vailes Por Hora')

# Mostrar el histograma
plt.show()
```

En la próxima grafica observamos un histograma de la columna víctimas de accidentes. podemos observar que la gran mayoría de las víctimas de los accidentes son conductores o pasajero de motos y peatones.

```python
plt.hist(df2['VICTIMA'], bins=8)  # 'bins' es el número de divisiones del histograma

#plt.xticks(range(min(df2['VICTIMA']), max(df2['VICTIMA'])+1))
# Agregar etiquetas y título
plt.xticks(rotation=45)
plt.xlabel('Victimas')
plt.ylabel('Numero De Siniestros')
plt.title('Histograma De Victimas')

# Mostrar el histograma
plt.show()
```

En el próximo histograma que relaciona el número de accidente según el genero de la persona y podemos llegar a la conclusión que los hombres son muchos mas propenso a sufrir una accidente que una mujer

```python
plt.figure(figsize=(4, 6))
plt.hist(df2[df2['SEXO']!='SD']['SEXO'], bins=2)  # 'bins' es el número de divisiones del histograma

#plt.xticks(range(min(df2['VICTIMA']), max(df2['VICTIMA'])+1))
# Agregar etiquetas y título
# plt.xticks(rotation=45)
plt.xlabel('Genero')
plt.ylabel('Numero De Victimas')
plt.title('Histograma De Acidentalidad Por Genero')

# Mostrar el histograma
plt.show()
```
